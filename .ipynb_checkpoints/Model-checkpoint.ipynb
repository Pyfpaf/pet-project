Как видим в пробеге автомобилей имеются значения, которые сильно отличаются от большинства значения. Построим для пробега график ящих с усами.

```python
plt.figure(figsize=(5, 5))
plt.boxplot(df['mileage'])
plt.title('График ящик с усами для пробега')
plt.show()
```

Приведем аномалии в пробеге (mileage) в диапазон интерквартильного размаха

```python
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    
    return boundaries

boundaries = calculate_outliers(df['mileage'])
boundaries
```

```python
df.loc[df['mileage'] > boundaries[1], 'mileage'] = round(boundaries[1])
plt.figure(figsize=(5, 5))
plt.boxplot(df['mileage'])
plt.title('График ящик с усами для пробега')
plt.show()
```

## 4. Modeling

### 4.1. Baseline

Загрузим обработанный датасет.

```python
df = pd.read_csv('data/df_prepared.csv')
df.head()
```

```python
X = df.drop('price', axis=1)
y = df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

#### 4.1.1. Linear regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

predicted_train_lr = lr.predict(X_train)
predicted_test_lr = lr.predict(X_test)
print(f'Значение RMSE для Линейной регрессии на тренировочной выборке {mean_squared_error(y_train, predicted_train_lr, squared=False)}')
print(f'Значение RMSE для Линейной регрессии на тестовой выборке {mean_squared_error(y_test, predicted_test_lr, squared=False)}')
print(f'Значение MAE для Линейной регрессии на тренировочной выборке {mean_absolute_error(y_train, predicted_train_lr)}')
print(f'Значение MAE для Линейной регрессии на тестовой выборке {mean_absolute_error(y_test, predicted_test_lr)}')
```

```python
scores = cross_val_score(lr, X, y, cv=5, scoring=make_scorer(mean_absolute_error))
print(f'MAE на каждом фолде: {scores}')
print(f'Среднее значение MAE: {np.mean(scores)}')
print(f'Среднеквадратичное отклонение MAE: {np.std(scores)}')
```

#### 4.1.2. Random forest regressor

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred_rf_train = rf.predict(X_train)
pred_rf_test = rf.predict(X_test)
print(f'Значение RMSE для Случайного леса на тренировочной выборке {mean_squared_error(y_train, pred_rf_train, squared=False)}')
print(f'Значение RMSE для Случайного леса на тестовой выборке {mean_squared_error(y_test, pred_rf_test, squared=False)}')
print(f'Значение MAE для Случайного леса на тренировочной выборке {mean_absolute_error(y_train, pred_rf_train)}')
print(f'Значение MAE для Случайного леса на тестовой выборке {mean_absolute_error(y_test, pred_rf_test)}')
```

```python
scores = cross_val_score(rf, X, y, cv=5, scoring=make_scorer(mean_absolute_error), n_jobs=-1)
print(f'MAE на каждом фолде: {scores}')
print(f'Среднее значение MAE: {np.mean(scores)}')
print(f'Среднеквадратичное отклонение MAE: {np.std(scores)}')
```

### 4.2. Tuning our models

#### 4.2.1. Linear regression

Попробуем применить регуляризаторы для линейной регрессии.

##### 4.2.1.1. Linear regression L1 (Lasso)

```python
coef_lasso = []

for i in range(-5, 5):
    alpha = 10**i
    lasso_reg = Lasso(alpha = alpha, random_state=123)
    lasso_reg.fit(X_train, y_train)

    predicted_train_lasso = lasso_reg.predict(X_train)
    predicted_test_lasso = lasso_reg.predict(X_test)

    coef_lasso.append(list(lasso_reg.coef_))
    print('-'*5, 'lambda = ', alpha, '-'*5)
    
    print(f'Значение RMSE для Lasso на тренировочной выборке: {mean_squared_error(y_train, predicted_train_lasso, squared=False)}')
    print(f'Значение RMSE для Lasso на тестовой выборке: {mean_squared_error(y_test, predicted_test_lasso, squared=False)}')
    print(f'Значение MAE для Lasso на тренировочной выборке: {mean_absolute_error(y_train, predicted_train_lasso)}')
    print(f'Значение MAE для Lasso на тестовой выборке: {mean_absolute_error(y_test, predicted_test_lasso)}')

    scores = cross_val_score(lasso_reg, X, y, cv=5, scoring=make_scorer(mean_absolute_error), n_jobs=-1)
    print(f'MAE на каждом фолде: {scores}')
    print(f'Среднее значение MAE: {np.mean(scores)}')
    print(f'Среднеквадратичное отклонение MAE: {np.std(scores)}')
```

##### 4.2.1.2. Linear regression L2 (Ridge)

```python
coef_ridge = []

for i in range(-5, 5):
    alpha = 10**i
    ridge_reg = Ridge(alpha = alpha, random_state=123)
    ridge_reg.fit(X_train, y_train)

    predicted_train_ridge = ridge_reg.predict(X_train)
    predicted_test_ridge = ridge_reg.predict(X_test)

    coef_ridge.append(list(ridge_reg.coef_))
    print('-'*5, 'lambda = ', alpha, '-'*5)
    
    print(f'Значение RMSE для Ridge на тренировочной выборке: {mean_squared_error(y_train, predicted_train_ridge, squared=False)}')
    print(f'Значение RMSE для Ridge на тестовой выборке: {mean_squared_error(y_test, predicted_test_ridge, squared=False)}')
    print(f'Значение MAE для Ridge на тренировочной выборке: {mean_absolute_error(y_train, predicted_train_ridge)}')
    print(f'Значение MAE для Ridge на тестовой выборке: {mean_absolute_error(y_test, predicted_test_ridge)}')

    scores = cross_val_score(ridge_reg, X, y, cv=5, scoring=make_scorer(mean_absolute_error), n_jobs=-1)
    print(f'MAE на каждом фолде: {scores}')
    print(f'Среднее значение MAE: {np.mean(scores)}')
    print(f'Среднеквадратичное отклонение MAE: {np.std(scores)}')
```

#### 4.2.2. Random forest regressor

##### 4.2.2.1. GridSearchCV

```python
rf_gs = RandomForestRegressor()
params = {
    'n_estimators': np.arange(10, 160, 20),
    'max_depth': np.arange(2, 25, 4),
    'max_features': ['sqrt', 'log2', None],
    # 'min_samples_leaf': np.arange(1, 11),
    # 'min_samples_split': np.arange(2, 11)
}

gs = GridSearchCV(rf_gs, params, cv=5, scoring=make_scorer(mean_absolute_error), n_jobs=-1)
gs.fit(X_train, y_train)
gs.best_estimator_
gs.best_score_
```

```python
gs.best_estimator_
```

##### 4.2.2.2. Подбор гиперпараметров вручную

Перебор с помощью класса GridSearchCV не дал результата. Попробуем подобрать вручную гиперпараметры для улучшения метрики MAE.

```python
rf = RandomForestRegressor(n_estimators=150, random_state=123)
scores = cross_val_score(rf, X, y, cv=5, scoring=make_scorer(mean_absolute_error), n_jobs=-1)
print(f'MAE на каждом фолде: {scores}')
print(f'Среднее значение MAE: {np.mean(scores)}')
print(f'Среднеквадратичное отклонение MAE: {np.std(scores)}')
```

Изменение прочих гиперпараметров ухудшало метрику МАЕ и среднеквадратичное отклонение
